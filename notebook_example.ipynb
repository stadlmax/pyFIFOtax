# Daily Exchange Rates
This is should also be legal (should not is, consult a tax advisor if in doubt). It can make a difference. If the difference is large enough, it might be worth a shot. 

Note that the **exact exchange rates** e.g. the ones you got on your wire transfers, are definitely **NOT** legal which has been tested in court.

As long as you use the EZB reference courses, it probably should be fine both ways. However, the monthly exchange rates might make it easier in terms of reporting gains/losses on foreign currencies, e.g. if you always have a wire transfer a few days after a sale order and might be more reliable with the Finanzamt.

```python
# analzye data for report year 2022 and use "daily" reference exchange rates
# the other alternative is "monthly_avg", i.e. per-month average rates (both EZB)
dfs_daily = report.consolidate_report(2022, "daily")
report.create_excel_report(2022, "daily", "tax_report_2022_daily_rates.xlsx")
```
